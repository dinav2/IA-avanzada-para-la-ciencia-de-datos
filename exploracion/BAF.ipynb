# Histogramas

¿Qué distribuciones (histogramas) tienen las variables numéricas?

```python
def plot_hist(df, column, bins=5, color='skyblue', title=None, xlabel=None):
    plt.figure(figsize=(6, 4))
    df[column].hist(bins=bins, color=color, edgecolor='black')
    plt.title(title if title else f"{column} Distribution")
    plt.xlabel(xlabel if xlabel else column)
    plt.ylabel("Frequency")
    plt.show()

plot_hist(base, 'proposed_credit_limit', bins=5, color='skyblue',
          title="Proposed Credit Limit Distribution", xlabel="Proposed Credit Limit")
```

```python
plot_hist(base, 'customer_age', bins=5, color='orange',
          title="Age Distribution", xlabel="Customer Age")
```
